# Search
We first encode our input features and take the predicted values to then be unraveled and flattened to feed into a DT. Note there feautres are 8 units in length and there are 6 observations and so the flattened features are 48 dim in size to be fed into the decision tree.

```python
encoded_features = []
for c in range(6):
    encoded_features.append(nn_encoder.predict(data[0][:,c,:,:,:], batch_size = 16)[2])
    tf.keras.backend.clear_session()
encoded_features = np.array(encoded_features)
```

```python
swapped_features = np.swapaxes(encoded_features, 0, 1)
total = []
for i in range(swapped_features.shape[0]):
    total.append(swapped_features[i,:,:].ravel())
total = np.array(total)
probs_list = forest.predict_proba(total)[:, 1]
print(probs_list)
```

```python
probs_list.max()
```

# Plot Top Result
Here we plot the top result. This gives us a candidate thats found in our nature astronomy paper `MLC8`.

```python
%matplotlib inline
best_arg = np.argsort(probs_list)[-1]

for i in range(6):
    plt.figure(figsize=(10,4))
    plt.xlabel("Fchans")
    plt.ylabel("Time")
    plt.imshow(data[0][best_arg,i,:,:,0], interpolation='nearest', cmap=plt.get_cmap('hot'))
    plt.savefig('test.png')
```
